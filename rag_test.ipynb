WEB  SCRAPPING

```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader(web_paths=["https://www.deeplearningbook.org/"])
docs = loader.load()
print(docs)
```

Splitting documents

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size = 1000, chunk_overlap = 200)
splits = text_splitter.split_documents(docs)
```

```python
print(splits[0])
print(splits[1])
```

```python
print(len(splits))
```

Add docs to VectorDB

```python
!pip install -U langchain langchain-community langchain-openai chromadb
```

```python
from langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import Chroma
vectorstore = Chroma.from_documents(documents=splits, embedding=OpenAIEmbeddings())

```

```python
print(vectorstore._collection.count())
```

```python
print(vectorstore._collection.get())
```

```python
print("\nCollection 1 - ", vectorstore._collection.count())
print(vectorstore._collection.get)
```

Rag pipeline

```python
retriever = vectorstore.as_retriever
```

Augmentation

```python
!pip install langchain-hub
```

```python
from langchain import hub
prompt = hub.pull("rlm/rag-prompt")
```

```python
try:
    import langchain_hub
    print('langchain_hub is installed successfully.')
except ImportError:
    print('langchain_hub is not installed.')
```

```python
from langchain_openai import ChatOpenAI
llm = ChatOpenAI()
```

```python
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser
```

```python
def format_docs(docs):
  return "\n".join(doc.page_content for doc in docs)
```

```python
rag_chain = ({"context" : retriever | format_docs, "question": RunnablePassthrough()}
              | prompt
              | llm
              | StrOutputParser())
```

```python
rag_chain.invoke("What is Deep leraning")
```
